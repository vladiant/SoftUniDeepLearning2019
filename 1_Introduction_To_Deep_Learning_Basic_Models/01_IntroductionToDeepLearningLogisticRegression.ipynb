```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from sklearn.datasets import load_iris

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

from tensorflow.keras.metrics import categorical_accuracy
```

# Intro to Deep Learning
## Logistic Regression

```python
iris_data = load_iris()
attributes, labels = iris_data.data, iris_data.target
```

```python
attributes.shape
```

```python
labels.shape
```

```python
# Check how many labels are unique
len(set(labels))
```

```python
labels_onehot = tf.one_hot(labels, depth=len(set(labels))).numpy()
```

```python
model = Sequential([
    Input(shape=(4, )),
    Dense(3, activation="sigmoid")
])
```

```python
model.summary()
```

```python
# model.compile(optimizer="adam", loss="categorical_crossentropy")
model.compile(optimizer="adam", loss=tf.keras.losses.categorical_crossentropy)
```

```python
model.fit(attributes, labels_onehot, epochs=500)
```

```python
predictions = model.predict(attributes)
```

```python
(model.predict_classes(attributes)==labels).sum()/len(labels)
```

```python
acc = categorical_accuracy(labels_onehot, predictions)
```

```python
print(tf.math.reduce_mean(acc).numpy())
```

```python
# Now it is Neural Network
model1 = Sequential([
    Input(shape=(4, )),
    Dense(3, activation="softmax")
])
```

```python
model1.summary()
```

```python
# model.compile(optimizer="adam", loss="sparse_categorical_crossentropy")
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               loss=tf.keras.losses.categorical_crossentropy,
               metrics = [categorical_accuracy])
```

```python
model1.fit(attributes, labels_onehot, epochs=500)
```
