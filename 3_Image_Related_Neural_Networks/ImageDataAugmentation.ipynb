```python
from numpy import expand_dims

from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator

from matplotlib import pyplot
```

```python
# Load the image
img = load_img("bird.jpg")
```

```python
# Convert to numpy array
data = img_to_array(img)
```

```python
# Expand dimensions to one sample
samples = expand_dims(data, 0)
```

```python
# Create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
```

```python
# Prepare iterator
it = datagen.flow(samples, batch_size=1)
```

```python
# Random transform generator
def get_transform(data):
    i = 0
    while True:
        i = i + 1
        yield datagen.get_random_transform(data, seed=i)
```

```python
print('Print transformation data')
gen_transf = get_transform(data)
for i in range(9):
    # Get transformation
    transformation = next(gen_transf)
    print(transformation)
    # Apply transformation
    transf_image = datagen.apply_transform(image, transformation)
    # Plot raw pixel data
    pyplot.imshow(transf_image)
    # Show the figure
    pyplot.show()
```
